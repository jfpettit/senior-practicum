```python
import numpy as np
import torch
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
import gym
from reinforce import mlp_policy, training_stuff
import matplotlib.pyplot as plt
```

```python
env = gym.make('CartPole-v0')
nA = env.action_space.n
policy = mlp_policy(nA)
optimizer = optim.Adam(policy.parameters(), lr=1e-2)
misc = training_stuff(.9, env, policy, optimizer)
misc.train_loop(True, 100)
```

```python
plt.plot(range(len(misc.ep_length)), misc.ep_length)
plt.title('Episode length over epochs')
plt.ylabel('Episode length')
plt.xlabel('Epochs')
plt.show()
```

```python
plt.plot(range(len(misc.ep_reward)), misc.ep_reward)
plt.title('Episode reward over epochs')
plt.ylabel('Episode reward')
plt.xlabel('Epochs')
plt.show()
```
